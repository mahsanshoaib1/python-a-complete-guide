Seaborn categorized the plots in three categories:
1. Plots to find the relation (relplot)  default type scatter plot
2. Plots to find the distribution (displot) default type histplot
3. Plots for categorical variables (catplot) default type stripplot

These three are not actual plots, they have different kind of plot, for example relplot has two kind of plots i.e scatter and line plot. By default relplot creates scatter plot. These three provides some extra benifits which individual plots don't.

Plots can be plotted independently, like typing sns.scatterplot()

```python
import seaborn as sns
```

```python
# applying the default theme
sns.set_theme()
# load dataset
tips = sns.load_dataset("tips")
#create a visualization
sns.relplot(data = tips, x = "total_bill", y = "tip", col = "time", hue = "smoker", style = "smoker", size = "size")
```

> col is being used here to split the displot 

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False)
)
```

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
)
```

### Linear regression

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

### Distributional representations

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
# kde drawed the line
# dinner ky waqt bill ki count aur lunch ky waqt bill ki count
```

> KDE is kernal desity estimation

```python
sns.displot(data=tips, kind="ecdf", x="total_bill", col="time", hue="smoker", rug=True)
```

## Plots for categorical data

```python
sns.catplot(data=tips, kind="swarm", x="day", y="total_bill", hue="smoker")
```

```python
sns.catplot(data=tips, kind="violin", x="day", y="total_bill", hue="smoker",  split=True)
```

the mean value and its confidence interval within each nested category via the bar plot

```python
sns.catplot(data=tips, kind="bar", x="day", y="total_bill", hue="smoker")
```

## Multivariate views on complex datasets

```python
penguins = sns.load_dataset("penguins")
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

```python
sns.pairplot(data=penguins, hue="species")
```

## Lower-level tools for building figures

```python
g = sns.PairGrid(penguins, hue="species", corner=True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color=".2")
g.map_lower(sns.scatterplot, marker="+")
g.map_diag(sns.histplot, element="step", linewidth=0, kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6))
```

```python
sns.set_theme(style="ticks", font_scale=1.25)
g = sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g",
    palette="crest", marker="x", s=100,
)
g.set_axis_labels("Bill length (mm)", "Bill depth (mm)", labelpad=10)
g.legend.set_title("Body mass (g)")
g.figure.set_size_inches(6.5, 4.5)
g.ax.margins(.15)
g.despine(trim=True)
```
