## How to create new columns derived from existing columns

````python
air_quality = pd.read_csv("air_quality_no2.csv", index_col=0, parse_dates=True) # index col convert first column into index and pase_dates make it time series data.
air_quality.head(7)
````

````python
air_quality["london_mg_per_cubic"] = air_quality["station_london"] * 1.882 # created a new column and added to the dataset.
air_quality.head()
````

````python
air_quality["ratio_paris_antwerp"] = (
    air_quality["station_paris"] / air_quality["station_antwerp"]
)
air_quality.head()
````

````python
#  rename the data columns
air_quality_renamed = air_quality.rename(columns={
    "station_antwerp": "BETR801",
        "station_paris": "FR04014",
        "station_london": "London Westminster",
})
air_quality_renamed.head()
````

````python
air_quality_renamed = air_quality_renamed.rename(columns=str.lower)
air_quality_renamed.head()
````

## How to calculate summary statistics

````python
air_quality["station_paris"].mean()
````

````python
air_quality[["station_paris", "station_london"]].median()
````

````python
air_quality[["station_paris", "station_london"]].describe()
````

````python
# Instead of the predefined statistics, specific combinations of aggregating statistics for given columns can be defined using the DataFrame.agg() method:
air_quality[["station_paris", "station_london"]].agg(("mean", "median", "min", "max"))
````

````python
air_quality.agg({
    "station_paris":("mean", "median", "min", "max"),
     "station_london": ("mean", "median", "min", "max")
    })
````

## Aggregating statistics grouped by category

````python

import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic.head()
````

````python
# What is the average age for male versus female Titanic passengers?
titanic[["sex", "age"]].groupby("sex").mean()
# the output is the mean of female and mean of male separately
````

````python
titanic.groupby("sex")["age"].mean()
````

````python
# titanic["fare"].groupby(["sex","who"]).mean()         # through an error
titanic.groupby(["sex", "pclass"])["fare"].mean()
````

````python
titanic [["fare", "class"]].groupby("class").median()
````

````python
titanic.groupby("sex").mean()
````

````python
titanic.groupby("sex").mean(numeric_only=True)
````

````python
titanic["pclass"].value_counts()
````

````python
titanic.groupby("pclass")["pclass"].count()
````

>Both size and count can be used in combination with groupby. Whereas size includes NaN values and just provides the number of rows (size of the table), 
>count excludes the missing values. In the value_counts method, use the dropna argument to include or exclude the NaN values.
