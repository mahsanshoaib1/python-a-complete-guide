## Creating dataframe in pandas


A DataFrame is a 2-dimensional data structure

```python
df = pd.DataFrame(
    {
        "Name":["Ahsan", "Abdullah", "Harmain", "Sheraz"],
        "Field": ["BioInformatics", "Computer Science", "BioInformatics", "BioInformatics"],
        "Age": [19, 23, 20, 20]
    }
)
```

```python
df
```

When using a Python dictionary of lists, the dictionary keys will be used as column headers and the values in each list as columns of the DataFrame.

Each Coloumn in a dataframe is called **Series**

```python
s = pd.Series([4,7,5,5], name = "semester")
s
```

When selecting a single column of a pandas DataFrame, the result is a pandas Series

```python
age = df["Age"] # accessing a coloumn from a dataframe 
age
```

## Filtering rows

### Using condition

```python
df1.head()
```

```python
less_30 = df1[df1["age"] < 30]
less_30.head()
```

**Note**: To select rows based on a conditional expression, use a condition inside the selection brackets

The condition inside the selection brackets titanic["Age"] > 35 checks for which rows the Age column has a value larger than 35:

```python
df1["age"]>30
```

```python
heights = df1[(df1["height"] <= 155 )| (df1["height"] >= 170)]
heights.head()
```

### isin([])

```python
weight = df1[df1["weight"].isin([70.0])]
weight
```

```python
weight70_75 = df1[df1["weight"].isin([70.0,75.0])]
weight70_75.head()
```

```python
age_no_na = df1[df1["age"].notna()]
```
